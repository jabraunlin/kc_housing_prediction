```python
data = pd.read_csv('kc_house_data.csv')
```

```python
y = data.price
```

```python
X = data[['bedrooms','bathrooms','sqft_living','yr_built','yr_renovated']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)
lin_pred = lin.predict(X_test)
mean_absolute_error(y_test, lin_pred)

```

```python
plt.scatter(y_test, y_test - lin_pred, alpha=0.3)
plt.hlines(0,0, np.max(y_test))
plt.title('Residuals')
plt.savefig('lin_resids.png')
```

```python
data.date = pd.to_datetime(data.date)
split_date = data.date.quantile(q=0.8)
X_train = data[data['date']<= split_date]
X_test = data[data['date'] > split_date]
y_train = X_train.price
y_test = X_test.price
```

```python
avg_price = X_train.groupby(('bedrooms','bathrooms','zipcode')).mean()['price'].reset_index().rename(columns={'price':'avg_price'})
```

```python
X_train = pd.merge(X_train,avg_price, how='left', on=['bedrooms','bathrooms','zipcode'])
X_test = pd.merge(X_test, avg_price, how='left',on=['bedrooms','bathrooms','zipcode'])
```

```python
X_test.avg_price.fillna(np.mean(X_train['price'].values),inplace=True)
```

```python
y_train = X_train.price
y_test = X_test.price
X_train = X_train.drop(columns=['price','date','id'])
X_test = X_test.drop(columns=['price','date','id'])
```

```python
def neighbors_prediction(n):
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train[['long','lat']],y_train)
    idx = knn.kneighbors(X_test[['long','lat']])[1]
    preds = []
    for neighbors in range(len(idx)):
        X_train_neighbors = X_train.iloc[idx[neighbors]]
        y_train_neighbors = y_train.iloc[idx[neighbors]]
        lin = LinearRegression()
        lin.fit(X_train_neighbors, y_train_neighbors)
        preds.append(lin.predict(X_test.iloc[neighbors].reshape(1,-1))[0])
    return mean_absolute_error(y_test,np.array(preds))
```

```python
errors = []
k_neighbors = [i for i in range(5,405)][::20]
for k in k_neighbors:
    errors.append(neighbors_prediction(k))
```

```python
plt.plot(k_neighbors[1:], errors[1:])
plt.ylabel('Mean Abs Error')
plt.xlabel('K Neighbors')
plt.title('Choosing an Optimal K')
plt.savefig('choosing_k.png')
```
