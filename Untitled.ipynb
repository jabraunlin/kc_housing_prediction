```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('kc_house_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data.price
```

```python
X = data[['bedrooms','bathrooms','sqft_living','yr_built','yr_renovated']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)
lin_pred = lin.predict(X_test)
np.sqrt(mean_squared_error(y_test, lin_pred))

```

```python
plt.scatter(y_test, y_test - lin_pred, alpha=0.3)
plt.hlines(0,0, np.max(y_test))
plt.title('Residuals')
```

```python
y = data.price
X = data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
avg_price = X_train.groupby(('bedrooms','bathrooms','zipcode')).mean()['price'].reset_index().rename(columns={'price':'avg_price'})
```

```python
X_train = pd.merge(X_train,avg_price, how='left', on=['bedrooms','bathrooms','zipcode'])
X_test = pd.merge(X_test, avg_price, how='left',on=['bedrooms','bathrooms','zipcode'])
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train[['long','lat']],X_train['price'])
train_pred = knn.predict(X_train[['long','lat']])
test_pred = knn.predict(X_test[['long','lat']])
X_train['neighbors'] = train_pred
X_test['neighbors'] = test_pred
```

```python
X_test.avg_price.fillna(np.mean(X_train['price'].values),inplace=True)
```

```python
X_train = X_train.drop(columns=['price','date'])
X_test = X_test.drop(columns=['price','date'])
```

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=7, n_jobs=-1)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
np.sqrt(mean_squared_error(y_test, pred))
```
